```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('src/datasets/selic_expectativas.csv')

df.head()
```

```python
df.describe()
```

```python
df['Data'] = pd.to_datetime(df['Data'])

# Gráfico de linha da mediana da Selic
plt.figure(figsize=(12, 6))
plt.plot(df['Data'], df['Mediana'], marker='o', color='darkblue')
plt.title('Evolução da Mediana da Taxa Selic (Expectativas de Mercado)')
plt.xlabel('Data')
plt.ylabel('Taxa Selic (%)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 5))


plt.subplot(1, 2, 1)
sns.histplot(df['Mediana'], bins=15, kde=True, color='purple')
plt.title('Distribuição da Mediana da Selic')


plt.subplot(1, 2, 2)
sns.boxplot(x=df['Mediana'], color='orange')
plt.title('Boxplot da Mediana da Selic')

plt.tight_layout()
plt.show()
```

### Conclusão da Análise da Taxa Selic (Expectativas de Mercado)

A análise dos dados de expectativas de mercado para a Taxa Selic revelou um comportamento concentrado, com a **mediana oscilando predominantemente entre 13% e 15%**, como evidenciado pelo histograma. O pico em torno de 14,75% indica que grande parte dos agentes de mercado esperavam uma manutenção da taxa nesse patamar ao longo do período observado.

O **boxplot confirma essa concentração**, mostrando uma distribuição assimétrica com **poucos outliers abaixo de 12%**, o que representa uma expectativa minoritária de cortes mais agressivos.

A **série temporal da mediana da Selic** destaca variações curtas em sequência, o que pode ser reflexo das diferentes instituições financeiras ajustando suas projeções com base nas decisões do Copom, indicadores econômicos ou falas do Banco Central.

Já a estatística descritiva mostra:

- Média: **13,54%**
- Mediana: **13,50%**
- Desvio padrão: **~1,09**, indicando uma **baixa dispersão**
- Mínimo de **11%** e máximo de **15%**, reforçando que **a expectativa do mercado permaneceu dentro de uma faixa bastante previsível**

Esses resultados evidenciam **relativa estabilidade nas projeções** e reforçam a importância da Taxa Selic como âncora de expectativas no sistema financeiro nacional.
