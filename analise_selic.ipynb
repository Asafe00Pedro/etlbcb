```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('src/datasets/selic_expectativas.csv')

df.head()
```

```python
df.describe()
```

```python
df['Data'] = pd.to_datetime(df['Data'])

# Grfico de linha da mediana da Selic
plt.figure(figsize=(12, 6))
plt.plot(df['Data'], df['Mediana'], marker='o', color='darkblue')
plt.title('Evoluo da Mediana da Taxa Selic (Expectativas de Mercado)')
plt.xlabel('Data')
plt.ylabel('Taxa Selic (%)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 5))


plt.subplot(1, 2, 1)
sns.histplot(df['Mediana'], bins=15, kde=True, color='purple')
plt.title('Distribuio da Mediana da Selic')


plt.subplot(1, 2, 2)
sns.boxplot(x=df['Mediana'], color='orange')
plt.title('Boxplot da Mediana da Selic')

plt.tight_layout()
plt.show()
```

### Concluso da Anlise da Taxa Selic (Expectativas de Mercado)

A anlise dos dados de expectativas de mercado para a Taxa Selic revelou um comportamento concentrado, com a **mediana oscilando predominantemente entre 13% e 15%**, como evidenciado pelo histograma. O pico em torno de 14,75% indica que grande parte dos agentes de mercado esperavam uma manuteno da taxa nesse patamar ao longo do perodo observado.

O **boxplot confirma essa concentrao**, mostrando uma distribuio assimtrica com **poucos outliers abaixo de 12%**, o que representa uma expectativa minoritria de cortes mais agressivos.

A **srie temporal da mediana da Selic** destaca variaes curtas em sequncia, o que pode ser reflexo das diferentes instituies financeiras ajustando suas projees com base nas decises do Copom, indicadores econmicos ou falas do Banco Central.

J a estatstica descritiva mostra:

- Mdia: **13,54%**
- Mediana: **13,50%**
- Desvio padro: **~1,09**, indicando uma **baixa disperso**
- Mnimo de **11%** e mximo de **15%**, reforando que **a expectativa do mercado permaneceu dentro de uma faixa bastante previsvel**

Esses resultados evidenciam **relativa estabilidade nas projees** e reforam a importncia da Taxa Selic como ncora de expectativas no sistema financeiro nacional.
